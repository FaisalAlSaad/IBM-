##### Pentagon Area:

```python
def pen(a):
    G = (1/4)*(5*(5+2*(5)**(1/2)))**(1/2)
    a = pow(a,2)
    result = G * a
    print('The area of the pentagon is ', str(round(result, 2)))
```

```python
pen(5)
```
